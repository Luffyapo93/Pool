```python
import matplotlib.pyplot as plt
import numpy as np
import math

# 1. definir los datos
x = np.linspace(-5, 5, 100) # valores de x
y = x**2 # valores de y para y = x²

# 2. crear el grafico
plt.figure(figsize=(8, 6)) # tamaño del grafico
plt.plot(x, y,color="blue", label="y = x²") # trasado de linea azul

# 3. Añadir elementos al grafico
plt.xlabel('Eje x')
plt.ylabel('Eje y')
plt.title('Visualizacion de l aecuacion y = x²')
plt.legend()
plt.grid(True)

# 4. Mostrar el grafico
plt.show()
```

```python
import numpy as np

# Definir la matriz de coeficientes (A) y el vector de resultados (b)
A = np.array([[-1, 2], [2, -4]])
b = np.array([4, 3])

# Resolver el sistema de ecuaciones
try:
    x = np.linalg.solve(A, b)
    print("La solución del sistema es:")
    print("x =", x[0])
    print("y =", x[1])
except np.linalg.LinAlgError as e:
    print(f"No se pudo resolver el sistema de ecuaciones. Error: {e}")
    print("Esto podría indicar que el sistema no tiene solución única (es inconsistente o dependiente).")
```

```python
import numpy as np

# Función para resolver sistemas de ecuaciones Ax = b
def resolver_sistema(A, b, ejercicio_nombre):
    print(f"\n--- {ejercicio_nombre} ---")
    try:
        x = np.linalg.solve(A, b)
        print("La solución del sistema es:")
        print("x =", x[0])
        print("y =", x[1])
    except np.linalg.LinAlgError as e:
        print(f"No se pudo resolver el sistema de ecuaciones para {ejercicio_nombre}. Error: {e}")
        print("Esto podría indicar que el sistema no tiene solución única (es inconsistente o dependiente).")

# =========================
# EJERCICIO 10
# =========================
A_10a = np.array([[2, 1], [3, -1]])
b_10a = np.array([10, -10])
resolver_sistema(A_10a, b_10a, "EJERCICIO 10a")

a_10b = np.array([[2, -4], [2, -4]])
b_10b = np.array([3, -2])
resolver_sistema(a_10b, b_10b, "EJERCICIO 10b")

# =========================
# EJERCICIO 11
# =========================
A_11 = np.array([[2, -1], [3, 1]])
b_11 = np.array([11, 5])
resolver_sistema(A_11, b_11, "EJERCICIO 11")

# =========================
# EJERCICIO 12
# =========================
A_12 = np.array([[2, -1], [3, 2]])
b_12 = np.array([9, -2])
resolver_sistema(A_12, b_12, "EJERCICIO 12")

# =========================
# EJERCICIO 13
# =========================
A_13 = np.array([[2, -1], [3, -2]])
b_13 = np.array([7, 2])
resolver_sistema(A_13, b_13, "EJERCICIO 13")

# =========================
# EJERCICIO 14
# =========================
A_14 = np.array([[2, -1], [3, 2]])
b_14 = np.array([11, -6])
resolver_sistema(A_14, b_14, "EJERCICIO 14")

# =========================
# EJERCICIO 15
# =========================
A_15 = np.array([[2, -1], [3, -1]])
b_15 = np.array([5, 4])
resolver_sistema(A_15, b_15, "EJERCICIO 15")

# =========================
# EJERCICIO 16
# Recta: 5x - 4y = 1 → soluciones infinitas
print("\n--- EJERCICIO 16 ---")
print("Recta: 5x - 4y = 1 → soluciones infinitas. Este es un sistema con una sola ecuación y dos variables, que tiene infinitas soluciones que se encuentran sobre la recta.")

# =========================
# EJERCICIO 17
# Recta: 3x - 2y = 1
print("\n--- EJERCICIO 17 ---")
print("Recta: 3x - 2y = 1. Similar al Ejercicio 16, este sistema con una sola ecuación y dos variables tiene infinitas soluciones sobre la recta.")

# =========================
# EJERCICIO 18
# Recta: 5x + 2y = 3
print("\n--- EJERCICIO 18 ---")
print("Recta: 5x + 2y = 3. Similar a los Ejercicios 16 y 17, este sistema con una sola ecuación y dos variables tiene infinitas soluciones sobre la recta.")

# =========================
# EJERCICIO 19
print("\n--- EJERCICIO 19 ---")
print("Ejercicio 19: Soluciones infinitas (ejemplos: (0,0), (2,1), (4,2)). Este parece referirse a una ecuación lineal homogénea o un sistema dependiente.")

# =========================
# EJERCICIO 20
print("\n--- EJERCICIO 20 ---")
print("Ejercicio 20: Comprobación de pares de soluciones en una ecuación dada. Esto implica sustituir los pares de valores en la ecuación para verificar si la satisfacen.")

# =========================
# EJERCICIO 21
print("\n--- EJERCICIO 21 ---")
print("Ejercicio 21: El sistema no tiene solución (rectas paralelas). Esto ocurre cuando las ecuaciones son inconsistentes.")


# =========================
# EJERCICIO 22
# =========================
A_22 = np.array([[5, -2], [2, 2]])
b_22 = np.array([2, 2])
resolver_sistema(A_22, b_22, "EJERCICIO 22")

# =========================
# EJERCICIO 23
print("\n--- EJERCICIO 23 ---")
print("Ejercicio 23: Sistema sin solución (rectas paralelas). Similar al Ejercicio 21, esto indica un sistema inconsistente.")

# =========================
# EJERCICIO 24
# =========================
A_24 = np.array([[2, -1], [2, -2]])
b_24 = np.array([2, -2])
resolver_sistema(A_24, b_24, "EJERCICIO 24")

# =========================
# EJERCICIO 25
# =========================
A_25 = np.array([[2, 1], [2, -1]])
b_25 = np.array([0, 4])
resolver_sistema(A_25, b_25, "EJERCICIO 25")
```
